## Array creation
>Create the followings to numpy array:
* A: a 3\*3 array like this:
$$
A = \left(\begin{array}{cc} 
10 & 20 & 30\\
5 & 10 & 15\\
1 & 5 & 8
\end{array}\right)
$$ 
* B: a 3\*3\*4 that filled with the numbers start from 15 and have steps of 5 to the end. The array should be filled first along the first dimension then the second dimension and at last the third dimension. like this:
$$
B[:, :, 0] = \left(\begin{array}{cc} 
15 & 30 & 45\\
20 & 35 & 50\\
25 & 40 & 55
\end{array}\right)
$$ 

```python
### START CODE HERE ###

A = np.array([[10, 20, 30], [5, 10, 15], [1, 5, 8]])
B = np.arange(15, 195, 5).reshape((4, 9))
B = B.reshape((4, 3, 3), order = 'F')

### TEST

print("A = %s" % A)
print("B = %s" % B)

### END CODE HERE ###
```

Fo creating B, first we arange numbers 15 to 190 with stepsize = 5. Then by using the function "reshape", we change its dimension to 4x9, and in the next step we reshape it to 3x3x4 but in "F" order which fills the matrix column by column. We could at very first step reshape the numbers to 3x3x4 but the order demanded in the question wouldn't be met.

## Transposing 
>Transpose the array B elements in each slice. Then print the result.

```python
### START CODE HERE ###

B = np.transpose(B, (0, 2, 1))
print(B)

### END CODE HERE ###
```

## Matrix Multiplication
>Calculate the matrix multiplication of all slices of the array B. 
(B[:, :, 0] \* B[:, :, 1] \* B[:, :, 2] \* B[:, :, 3]). Then print the result.

>**Note**: the result should be a 3\*3 array. 

```python
### START CODE HERE ###

C = np.matmul(B[0, :, :], B[1, :, :])
C = np.matmul(C, B[2, :, :])
C = np.matmul(C, B[3, :, :])
print(C)

### END CODE HERE ###
```

## Actors/Actresses Activity
>For each Actors/Actresses caculate the number of movies in the dataframe, then sort the actors/actresses first by their number of movies in descending order and then by their name in ascending order. 

```python
### START CODE HERE ###
    
actors = df['Actors']
actors = [actors[i].split(',') for i in range(0, 1000)]
actors =  [actor for sublist in actors for actor in sublist]
actors = [actors[i].strip() for i in range(0, len(actors))]
ACTORS = set(actors)

movie_numbers = []

for actor in ACTORS:
    n = actors.count(actor)
    movie_numbers.append(n)
    
activity = dict(zip(ACTORS, movie_numbers))
    
### TEST

activity_1 = sorted(activity.items(), key=lambda item: item[1], reverse=True)
print("Actors/Actresses sorted by their numbers of movies in descending order: %s" % activity_1)

activity_2 = sorted(activity.items(), key=lambda item: item[0])
print("Actors/Actresses sorted by their names in ascending order: %s" % activity_2)

### END CODE HERE ###
```

Firstly, we seperate the data of actors' column. Then by using "split" we seperate the actors of a single movie and store all of them in a list. Secondly, by using "set" we remove repeated elements and store it in a new array. Then, we count the number of movies for each element of the new array in the old one and store them. In the end, we create a dict from these recent arrays using "zip". Finally, we sort them based on the order demanded. Also the name of the actors consist of some spaces at the beginning and the end. Using "strip" function will remove them.

## Plot
>You must plot the number of movies in each year according to the IMDB-Movie-Data.csv file. X-axis is year and y-axis show the number of movies in each year.
> set the x_label, y_label and title for your plot.

```python
### START CODE HERE ###

years = df['Year']
years = sorted(years)
YEARS = set(years)
YEARS = sorted(YEARS)

diagram = []
d = 0

for i in range(0, len(YEARS)):
    n = years.count(years[d])
    d = d + n
    
    diagram.append(n)
    
plt.plot(YEARS, diagram)
plt.title("The number of movies in each year according to the IMDB-Movie-Data.csv file")
plt.xlabel("year")
plt.ylabel("number of movies")
plt.show()
    
### END CODE HERE ###
```

## Plot
>You must plot the cos(x) curve. plot 20 cycle of the curve. the frequency of first half of curve is 50 Hz and frequency of second half is 200 Hz. 
> set the x_label, y_label and title for your plot.

```python
### START CODE HERE ###

x1 = np.linspace(0, 80, 1000)
x2 = np.linspace(80, 160, 1000)
f1 = 50
f2 = 200
y1 = np.cos(2*np.pi*f1*x1)
y2 = np.cos(2*np.pi*f2*x2)


plt.plot(x1, y1, 'g')
plt.plot(x2, y2, 'g')
plt.title("20 cycles of the cosine curve with frequency of first half = 50 Hz and frequency of second half = 200 Hz")
plt.xlabel("x")
plt.ylabel("cos 50Hz and 200Hz")
plt.show()

### END CODE HERE ###
```
